```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
boltzmann = 1.38064852e-23
```

```python
f50 = open("50averages.dat")
f100= open("30averages.dat")
f150 = open("70averages.dat")
t_50, m_50, m2_50, e_50, e2_50 =[], [], [], [], []
t_100, m_100, m2_100, e_100, e2_100 =[], [], [], [], []
t_150, m_150, m2_150, e_150, e2_150 =[], [], [], [], []

for l in f50:
    row = l.split()
    t_50.append(float(row[0]))
    m_50.append(abs(float(row[1])))
    m2_50.append(float(row[2]))
    e_50.append(float(row[3]))
    e2_50.append(float(row[4]))
for l in f100:
    row = l.split()
    t_100.append(float(row[0]))
    m_100.append(abs(float(row[1])))
    m2_100.append(float(row[2]))
    e_100.append(float(row[3]))
    e2_100.append(float(row[4]))
for l in f150:
    row = l.split()
    t_150.append(float(row[0]))
    m_150.append(abs(float(row[1])))
    m2_150.append(float(row[2]))
    e_150.append(float(row[3]))
    e2_150.append(float(row[4]))
```

```python
chi_50  = [abs(i*i - j) for (i,j) in zip(m_50,m2_50)]
cv_50   = [abs(i*i - j) for (i,j) in zip(e_50,e2_50)]
chi_100 = [abs(i*i - j) for (i,j) in zip(m_100,m2_100)]
cv_100  = [abs(i*i - j) for (i,j) in zip(e_100,e2_100)]
chi_150 = [abs(i*i - j) for (i,j) in zip(m_150,m2_150)]
cv_150  = [abs(i*i - j) for (i,j) in zip(e_150,e2_150)]
```

```python
plt.xlabel('Temperature')
plt.ylabel('m(T)')
plt.plot(t_50,m_50, label = '50x50')
plt.plot(t_100,m_100, label ='30x30')
plt.plot(t_150,m_150, label= '70x70')
plt.legend(loc='best')
plt.show()
```

```python
plt.xlabel('Temperature')
plt.ylabel('E(T)')
plt.plot(t_50,e_50, label = '50x50')
plt.plot(t_100,e_100, label ='30x30')
plt.plot(t_150,e_150, label= '70x70')
plt.legend(loc='best')
plt.show()
```

```python
plt.xlabel('Temperature')
plt.ylabel('chi(T) magnetic susceptibility')
plt.plot(t_50,chi_50, label = '50x50')
plt.plot(t_100,chi_100, label ='30x30')
plt.plot(t_150,chi_150, label= '70x70')
plt.legend(loc='best')
plt.show()
```

```python
plt.xlabel('Temperature')
plt.ylabel('cv(T) specific heat')
plt.plot(t_50,cv_50, label = '50x50')
plt.plot(t_100,cv_100, label ='30x30')
plt.plot(t_150,cv_150, label= '70x70')
plt.legend(loc='best')
plt.show()
```

```python
t_50[0]+=0.0001
t_100[0]+=0.0001
t_150[0]+=0.0001
chi_50_t  = [abs(i*i - j)/(t*boltzmann) for (i,j,t) in zip(m_50,m2_50,t_50)]
cv_50_t   = [abs(i*i - j)/(t*t*boltzmann) for (i,j,t) in zip(e_50,e2_50,t_50)]
chi_100_t = [abs(i*i - j)/(t*boltzmann) for (i,j,t) in zip(m_100,m2_100,t_100)]
cv_100_t  = [abs(i*i - j)/(t*t*boltzmann) for (i,j,t) in zip(e_100,e2_100,t_100)]
chi_150_t = [abs(i*i - j)/(t*boltzmann) for (i,j,t) in zip(m_150,m2_150,t_150)]
cv_150_t  = [abs(i*i - j)/(t*t*boltzmann) for (i,j,t) in zip(e_150,e2_150,t_150)]
```

```python
plt.xlabel('Temperature')
plt.ylabel('chi(T) magnetic susceptibility/t')
plt.plot(t_50,chi_50_t, label = '50x50')
plt.plot(t_100,chi_100_t, label ='30x30')
plt.plot(t_150,chi_150_t, label= '70x70')
plt.legend(loc='best')
plt.show()
```

```python
plt.xlabel('Temperature')
plt.ylabel('cv(T) specific heat/t')
plt.plot(t_50,cv_50_t, label = '50x50')
plt.plot(t_100,cv_100_t, label ='30x30')
plt.plot(t_150,cv_150_t, label= '70x70')
plt.legend(loc='best')
plt.show()
```
