```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
f50 = open("results_50_30.dat")
f100 = open("results_100_30.dat")
f150 = open("results_150_30.dat")
p_50, m_50, ps_50 =[], [], []
p_100, m_100, ps_100 =[], [], []
p_150, m_150, ps_150 =[], [], []

for l in f50:
    row = l.split()
    p_50.append(float(row[0]))
    ps_50.append(float(row[1]))
    m_50.append(float(row[2]))
for l in f100:
    row = l.split()
    p_100.append(float(row[0]))
    ps_100.append(float(row[1]))
    m_100.append(float(row[2]))
for l in f150:
    row = l.split()
    p_150.append(float(row[0]))
    ps_150.append(float(row[1]))
    m_150.append(float(row[2]))
```

```python
plt.xlabel('p')
plt.ylabel('maximum size of cluster')
plt.plot(p_50,m_50, label = '50x50')
plt.plot(p_100,m_100, label ='100x100')
plt.plot(p_150,m_150, label= '150x150')
plt.legend(loc='best')
plt.show()
```

```python
plt.xlabel('p')
plt.ylabel('probability of scanning cluster')
plt.plot(p_50,ps_50, label = '50x50')
plt.plot(p_100,ps_100, label ='100x100')
plt.plot(p_150,ps_150, label= '150x150')
plt.legend(loc='best')
plt.show()
```
